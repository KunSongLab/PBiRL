```python
import folium
import json
import geopandas as gpd
```

```python
with open("controllable_agent_info.json", "r") as f:
    agent_info = json.load(f)
var = []
node_id = []
for agent_id in agent_info.keys():
    var.append(agent_info[agent_id]["location"])
    node_id.append(agent_info[agent_id]["node_id"])
```

```python


m = folium.Map(location=[30.740749704487694,104.18661929844393], tiles='cartodbpositron', zoom_start=12)

controlist = [int(agent_id) for agent_id in agent_info.keys()]
# controlist = [28, 45, 39, 46, 56, 58, 102, 133, 135, 137, 141] + list(range(144, 172))
# polylist = [each - 144 for each in controlist if each >= 144]
# region_poly = region_poly.iloc[polylist]

all_agent_id = [int(agent_id) for agent_id in agent_info.keys()]
for i, each in enumerate(var):
    if all_agent_id[i] not in controlist:
        continue

    # if i < 144:
    if node_id[i] > 0:
        folium.Marker(
            location=[each[1], each[0]],
            radius=20,
            color="#d53e4f",
            popup=all_agent_id[i] - 1,
            icon=folium.Icon(color='red', icon='info-sign')
        ).add_to(m)
    else:
        folium.Circle(
            location=[each[1], each[0]],
            radius=20,
            color="blue",
            popup=all_agent_id[i]
        ).add_to(m)

m.save("RingFreeway-1.html")
```
